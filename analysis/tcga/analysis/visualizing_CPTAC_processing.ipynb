```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# processing CPTAC data
```

```python
file="/mnt/f/TCGA_junctions/ext_dat/CPTAC/kmers_in_cptac.txt"
cptac_data = pd.read_table(file,sep="\t",header=None,names=["splicemutr_kmers","cptac_kmers","percentage"])
```

```python
percentages=cptac_data["percentage"].tolist()
```

```python
plt.boxplot(percentages)
plt.ylabel("CPTAC kmers/all immunogenic kmers")
```

```python
plt.bar(list(range(len(percentages))),percentages)
plt.ylabel("CPTAC kmers/all immunogenic kmers")
plt.xlabel("Sample")
```
